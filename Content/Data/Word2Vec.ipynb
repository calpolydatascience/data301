Remove punctuation and then remove empty strings. This will delete items in the list that were previously just punctuation.

```python
import string
words = [''.join(c for c in s if c not in string.punctuation)
         for s in words]
words = [s for s in words if s]
print(len(words))
words
```

Remove common words ("for", "a", "of", "the", "and", "to", "in") by changing them into empty strings and then remove empty strings.

```python
stoplist = set('for a of the and to in'.split())
texts = [[word for word in item.lower().split() if word not in stoplist] for item in words]
texts = [s for s in texts if s]
print(len(words))
print(len(texts))
texts
```

Remove words that only appear once by changing them into empty strings and then remove empty strings.

```python
from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1
texts = [[token for token in text if frequency[token] > 1]
         for text in texts]
texts = [s for s in texts if s]
print(len(texts))
texts
```

Save the dictionary for future reference.

```python
from gensim import corpora
dictionary = corpora.Dictionary(texts)
dictionary.save('/tmp/deerwester.dict')
print(dictionary)
```

Give each word that appears in the document a unique id with the _token2id_ method.

```python
print(dictionary.token2id)
```

The _doc2bow_ method takes the string of words in the object and returns a list of 2-tuples. Each 2-tuple contains the corresponding id of the word followed by a 1 if the word appears in the document. By default, return_missing is set to false, in which case the words in the string that don't appear in the document are not produced in the output. By setting return_missing to true, the words not in the document are also shown separately.

```python
new_doc = "fist fish cows blasphemous however"
new_vec = dictionary.doc2bow(new_doc.lower().split(), return_missing=True)
print(new_vec) # the word "interaction" does not appear in the dictionary and is ignored
```
